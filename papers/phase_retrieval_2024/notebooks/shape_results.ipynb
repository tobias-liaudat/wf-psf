```python
import numpy as np

import matplotlib.ticker as mtick
import wf_psf as wf
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
base_path = '/Users/ec270266/Documents/Phd/Euclid/dev/phase_retrieval/wf-psf/papers/phase_retrieval_2024/data/paper_output_v1/'
log_folder = 'log-files/'
metric_base_path = '/Users/ec270266/Documents/Phd/Euclid/dev/phase_retrieval/wf-psf/papers/phase_retrieval_2024/data/paper_output_v1/full_metrics/'

metrics_base_id_full = 'metrics-poly_wf_PR_NewPRoj_12_cycles_full_eval_v'
metrics_base_id_param = 'metrics-poly_wf_PR_NewPRoj_12_cycles_onlyparam_eval_v'
    
```

# Shape results

```python
cycles = ['12']
runs = ['0', '2', '3']

e1_results_full = np.zeros((len(cycles), len(runs)))
e1_results_full_std = np.zeros((len(cycles), len(runs)))
e2_results_full = np.zeros((len(cycles), len(runs)))
e2_results_full_std = np.zeros((len(cycles), len(runs)))
R_results_full = np.zeros((len(cycles), len(runs)))
R_results_full_std = np.zeros((len(cycles), len(runs)))

e1_results_param = np.zeros((len(cycles), len(runs)))
e1_results_param_std = np.zeros((len(cycles), len(runs)))
e2_results_param = np.zeros((len(cycles), len(runs)))
e2_results_param_std = np.zeros((len(cycles), len(runs)))
R_results_param = np.zeros((len(cycles), len(runs)))
R_results_param_std = np.zeros((len(cycles), len(runs)))

for cycle_idx, cycle_ in enumerate(cycles):
    for run_idx, run_ in enumerate(runs):
        loaded_metrics_full = np.load(
            metric_base_path + metrics_base_id_full + run_ + '_' + 'cycle' + cycle_ + '.npy',
            allow_pickle=True
        )[()]
        loaded_metrics_param = np.load(
            metric_base_path + metrics_base_id_param + run_ + '_' + 'cycle' + cycle_ + '.npy',
            allow_pickle=True
        )[()]
        e1_results_param[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['rmse_e1']
        e1_results_full[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e1']
        e2_results_param[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['rmse_e2']
        e2_results_full[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e2']
        R_results_param[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['rmse_R2_meanR2']
        R_results_full[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_R2_meanR2']

        e1_results_full_std[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_e1']
        e2_results_full_std[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_e2']
        R_results_full_std[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_R2_meanR2']
        e1_results_param_std[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['std_rmse_e1']
        e2_results_param_std[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['std_rmse_e2']
        R_results_param_std[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['shape_results_dict']['std_rmse_R2_meanR2']


# Original metric
metrics_base_path = '/Users/ec270266/Documents/Phd/Euclid/dev/phase_retrieval/wf-psf/papers/phase_retrieval_2024/data/paper_output_v1/metrics/'

e1_results_original = np.zeros(len(runs))
e2_results_original = np.zeros(len(runs))
R_results_original = np.zeros(len(runs))
e1_results_original_std = np.zeros(len(runs))
e2_results_original_std = np.zeros(len(runs))
R_results_original_std = np.zeros(len(runs))

control_e1 = np.zeros(len(runs))
control_e2 = np.zeros(len(runs))
control_R = np.zeros(len(runs))

for run_idx, run_ in enumerate(runs):
    loaded_metrics_full = np.load(
            metrics_base_path + 'metrics-poly_wf_PR_NewPRoj_12_cycles_v' + run_ + '.npy',
            allow_pickle=True
        )[()]
    control_e1[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e1']
    control_e2[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e2']
    control_R[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_R2_meanR2']

    loaded_metrics_full = np.load(
            metrics_base_path + 'metrics-poly_wf_PR_benchmark_2_cycles_v' + run_ + '.npy',
            allow_pickle=True
        )[()]
    e1_results_original[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e1']
    e2_results_original[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_e2']
    R_results_original[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['rmse_R2_meanR2']
    e1_results_original_std[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_e1']
    e2_results_original_std[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_e2']
    R_results_original_std[run_idx] = loaded_metrics_full['test_metrics']['shape_results_dict']['std_rmse_R2_meanR2']
```

```python
scale = 1e3
decimals = 1
print("e1 results [{}]:".format(scale))
print('full:    ', np.around(e1_results_full * scale, decimals), '+/-', np.around(e1_results_full_std * scale, decimals))
print('param:   ', np.around(e1_results_param * scale, decimals), '+/-', np.around(e1_results_param_std * scale, decimals))
print('original:', np.around(e1_results_original * scale, decimals), '+/-', np.around(e1_results_original_std * scale, decimals))
print('control: ', np.around(control_e1 * scale, decimals))
```

```python
print("e2 results [{}]:".format(scale))
print('full:    ', np.around(e2_results_full*scale, decimals) , '+/-', np.around(e2_results_full_std*scale, decimals))
print('param:   ', np.around(e2_results_param*scale, decimals), '+/-', np.around(e2_results_param_std*scale, decimals))
print('original:', np.around(e2_results_original*scale, decimals), '+/-', np.around(e2_results_original_std*scale, decimals))
print('control: ', np.around(control_e2*scale, decimals))
```

```python
print("Size results R:")
print('full:    ', np.around(R_results_full*scale, decimals) , '+/-', np.around(R_results_full_std*scale, decimals))
print('param:   ', np.around(R_results_param*scale, decimals), '+/-', np.around(R_results_param_std*scale, decimals))
print('original:', np.around(R_results_original*scale, decimals), '+/-', np.around(R_results_original_std*scale, decimals))
print('control: ', np.around(control_R*scale, decimals))
```
