# WFE results

```python
cycles = ['2', '5', '9', '12']
runs = ['0', '2', '3']

opd_results_full = np.zeros((len(cycles), len(runs)))
opd_results_param = np.zeros((len(cycles), len(runs)))

opd_std_full = np.zeros((len(cycles), len(runs)))
opd_std_param = np.zeros((len(cycles), len(runs)))

for cycle_idx, cycle_ in enumerate(cycles):
    for run_idx, run_ in enumerate(runs):
        loaded_metrics_full = np.load(
            metric_base_path + metrics_base_id_full + run_ + '_' + 'cycle' + cycle_ + '.npy',
            allow_pickle=True
        )[()]
        loaded_metrics_param = np.load(
            metric_base_path + metrics_base_id_param + run_ + '_' + 'cycle' + cycle_ + '.npy',
            allow_pickle=True
        )[()]
        opd_results_param[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['opd_metric']['rel_rmse_opd']
        opd_results_full[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['opd_metric']['rel_rmse_opd']
        opd_std_param[cycle_idx, run_idx] = loaded_metrics_param['test_metrics']['opd_metric']['rel_rmse_std_opd']
        opd_std_full[cycle_idx, run_idx] = loaded_metrics_full['test_metrics']['opd_metric']['rel_rmse_std_opd']
```

```python
def argmedian(a, axis=0, keepdims=False):
    # Return the indices of the median values along an axis.
    a = np.asarray(a)
    idx = np.argsort(a, axis=axis)
    mid = a.shape[axis] // 2
    if a.shape[axis] % 2 == 0:
        mid -= 1
    arg_median = np.take(idx, mid, axis=axis)
    if keepdims:
        arr_median = np.zeros_like(a)
        for i in range(a.T.shape[axis]):
            arr_median[i, arg_median[i]] = 1
        return arr_median
    return arg_median
```

```python
print("OPD resutls only param:")
print(opd_results_param)
print("Median:")
print(np.median(opd_results_param, axis=1))
print("Best:")
print(np.min(opd_results_param, axis=1))
print()
print("std:")
print(opd_std_param)

argmedian(opd_results_param, axis=1, keepdims=True)
```

```python
print("OPD resutls full:")
print(opd_results_full)
print("Median:")
print(np.median(opd_results_full, axis=1))
print("Best:")
print(np.min(opd_results_full, axis=1))
print()
print("std:")
print(opd_std_full)

argmedian(opd_results_full, axis=1, keepdims=True)
```
