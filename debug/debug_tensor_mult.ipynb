```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%pylab inline
```

```python


pi = np.random.rand(2,21)
tf_pi = tf.constant(pi, dtype=tf.float32)

alpha = np.random.rand(21,21)
tf_alpha = tf.constant(alpha, dtype=tf.float32)

S = np.random.rand(21, 64, 64)
tf_S = tf.constant(S, dtype=tf.float32)

```

```python



inter_res = tf.linalg.matmul(tf_pi, tf_alpha)
result_original = tf.tensordot(inter_res, tf_S, axes=1)
result_original.shape
```

```python


inter_res_v2 = tf.tensordot(tf_alpha, tf_S, axes=1)
inter_res_v2.shape
```

```python
result_v2 = tf.tensordot(tf_pi, inter_res_v2, axes=[[1],[0]])
result_v2.shape
```

```python
tf.reduce_sum(result_original - result_v2)
```

```python
a = np.random.rand(21,21)
tf_a = tf.constant(a, dtype=tf.float32)

b = np.random.rand(21,21)
tf_b = tf.constant(b, dtype=tf.float32)

Z_k = np.random.rand(21, 64, 64)
tf_Z_k = tf.constant(Z_k, dtype=tf.float32)
```

```python

t1 = tf.tensordot(tf.tensordot(tf_a, tf_b, axes=[[1],[0]]), tf_Z_k, axes=[[1],[0]])
```

```python
t2 = tf.tensordot(tf_a, tf.tensordot(tf_b, tf_Z_k, axes=[[1],[0]]), axes=[[1],[0]])
```

```python
tf.reduce_sum(t1 - t2)
```
